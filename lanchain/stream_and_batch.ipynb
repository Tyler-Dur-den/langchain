```python
import os
from dotenv import load_dotenv
load_dotenv()
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("groq:qwen/qwen3-32b")
```

Streaming

Model make take a lot of time to genrated rather than just giving the output at end when we use function stream it starts printing output while being generated not yet finished.

```python
model.invoke("give me some facts about pandas")
```

```python
for chunk in model.stream("give me some facts about pandas"):
    print(chunk.text, end="|", flush=True)
```

Batching

It is used when we have many independent requests to a model we can batch them together and calls many api calls at the same time(like I am using groq it has about 25 api calls per minute, now there is risk of running out of limit as many can happen at the same time so we use concurrency which sets the limit to number of api calls). As it does not depend on each other so the second request can be done before completion of first and this improves performance and reduces cost because they are indepent we it does not have history or less token usage.

```python
responses = model.batch([
    "Give me 5 facts about pandas",
    "How do fish sleep?",
    "What is quantum computing?"
],
    config={
        'max_concurrency': 5
    }
    )

for response in responses:
    print(response)
```
